```python
df = pd.read_csv('bank-additional-full_normalised.csv')
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assuming df is your DataFrame with transaction data
# Preprocess data (example for scaling)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop(['class'], axis=1))

# Train Isolation Forest
clf = IsolationForest(n_estimators=100, contamination='auto')
clf.fit(scaled_features)

# Predict anomalies
df['anomaly'] = clf.predict(scaled_features)
```

```python
df['anomaly'] = df['anomaly'].map({-1: 1, 1: 0})
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(df['class'], df['anomaly']))
print(confusion_matrix(df['class'], df['anomaly']))
```

```python
# Get anomaly scores
df['scores'] = clf.decision_function(scaled_features)

# Examine the highest scoring anomalies
anomalies = df.sort_values(by='scores').head(20)
```
